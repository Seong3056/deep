- signup.csv 와 professor.csv 데이터를 이용하여 아래의 지시사항에 답하시오
1. signup.csv(수강데이터) 와 professor.csv(교수데이터) 를 각각 데이터프레임으로 불러오시오
 

2. 두 데이터프레임을 결합하시오

- 구성 컬럼 : 수강코드, 과목명, 담당교수, 학과
 

3. 각 학과별 담당교수 인원수를 출력하시오

```python
df = pd.read_csv("./dat/signup.csv", encoding='CP949')
df
```

```python
df1 = pd.read_csv("./dat/professor.csv", encoding='CP949')
df1
```

```python
df2 = pd.merge(df, df1, on = "담당교수", how = "inner")
df2
```

```python
grouped = df2.groupby(["학과"])
```

```python
counts = grouped.count()
counts["담당교수"]
```

```python
df2.groupby(["학과"]).count()["담당교수"]
```
