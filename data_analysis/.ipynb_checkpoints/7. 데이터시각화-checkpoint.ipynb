# 데이터시각화

- 데이터를 분석한 결과를 쉽게 이해할 수 있도록 표현하여 전달하는 것
- 데이터의 시각화가 필요한 이유
    - 많은 양의 데이터를 한 눈에 살펴볼 수 있음
    - 전문 지식이 없더라도 누구나 쉽게 데이터를 인지하고 사용할 수 있음
    - 단순한 데이터의 요약이나 나열보다 더 정확한 데이터 분석 결과를 얻을 수 있음
    - 단순한 데이터에서는 알 수 없었던 중요한 정보를 파악할 수 있음
    
- 파이썬의 시각화 라이브러리
    - matplotlib
        - 파이썬에서 가장 많이 사용하는 시각화 라이브러리
        - 판다스의 데이터프레임을 바로 시각화할 때도 내부적으로 matplotlib을 사용
        - 데이터 분속 이전의 데이터의 이해를 위해 시각화 또는 데이터 분석 이후의 결과를 시각화하기 위해 사용
        
    - seaborn
        - matplotlib을 기반으로 색 테마, 차트 기능 등을 추가해주는 라이브러리
        - matplotlib과 함께 사용하며 히트맵, 카운트플롯 등을 제공
        
    - folium
        - 지도 데이터를 이용하여 위치 정보를 시각화하는 라이브러리
        - 자바스크립트 기반의 인터렉티브 그래프를 그릴 수 있음
        
        
    - plotly(수업에서 안배움)
        - 인터렉티브 그래프를 그려주는 라이브러리
        
    - wordcloud(수업에서 안배움)

1. 시각화 라이브러리 불러오기

# 격자 설정

```python
plt.plot(a, a, "bo")
plt.plot(a, a**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(a, a**3, color = "springgreen", marker = "^", markersize = 9)
plt.grid(True, axis = "y", color = "red", alpha = 0.5, linestyle = "--")

plt.show()
```

# 눈금 표시

```python
plt.plot(a, a, "bo")
plt.plot(a, a**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(a, a**3, color = "springgreen", marker = "^", markersize = 9)

plt.xticks([0, 1, 2], labels = ["Jan", "Fed", "Mar"])
plt.yticks(np.arange(1, 6))

plt.tick_params(axis = "x", direction = "in", length = 13, pad = 6, labelsize = 14,
               labelcolor = "green", top = True)

plt.tick_params(axis = "y", direction = "inout", length = 10, pad =15,
                labelsize = 12, width = 2, color = "r")

plt.show()
```

- tick_params() : 눈금의 스타일을 설정
- axis : 설정이 적용될 축을 지정 x, y, both 중 선택
- direction : 눈금이 안에서 표시될지 밖에서 표시될지. in, out, inout
- length : 눈금의 길이
- pad : 눈금과 레이블과의 거리
- labelsize : 레이블의 크기
- labelcolor : 레이블의 색상
- top/bottom/left/right : 눈금이 표시될 위치 선택
- width : 눈금의 너비
- color : 눈금의 색상

# 막대 그래프

- 집단별 차이를 표현할 때 주로 사용
    - 크고 작음의 차이를 한 눈에 파악하기 위해
    - 예) 수량의 많고 적음 비교, 변화된 양에 대한 일별, 월별 연별 통계 등의 자료
    
- 가독성 면에서 막대 그래프는 일반적으로 항목의 개수가 적으면 가로 막대가 보기 좋고, 항목이 많으면 세로 막대가 보기 편함

```python
x = np.arange(3)
years = ["2022", "2023", "2024"]
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```

```python
x = np.arange(3)
years = ["2022", "2023", "2024"]
values = [100, 400, 900]

plt.bar(x, values, width = 0.6, align = "edge", color = "springgreen",
        edgecolor = "gray", linewidth = 3, tick_label = years, log = True)
plt.show()
```

- width : 막대의 너비
- align : 틱과 막대의 위치, edge로 설정하면 왼쪽 끝에 표시
- color : 막대의 색
- edgecolor : 막대의 테두리 색
- linewidth : 테두리의 두께
- log : y축이 로그 스케일로 표시

### 그룹 세로 막대 그래프

- 2016년 ~ 2019년 데이터를 막대 그래프로 그리기

```python
index = np.arange(4)

plt.figure(figsize = (15, 4))

data_bar = data.loc["Gyeonggi":"Daegu", 2016:2019]
for year in range(2016, 2020):
    chartdata = data_bar[year]
    plt.bar(index, chartdata, width = 0.2, label = year)
    index = index + 0.2
    
plt.ylim(35, 55)
plt.xlabel("area")
plt.ylabel("micrometer")
plt.xticks(index - 0.5, ["Gyeonggi", "Incheon", "Busan", "Daegu"])
plt.title("2016 ~ 2019 Fid Dust Bar Graph")
plt.legend()
plt.show()
```

# 산점도

- 서로 다른 두 연속형 변수 사이의 관계를 나타내며 연속형 변수의 상관성을 확인할 때 산점도 그래프를 사용
    - 예) 나이와 소득에 대한 상호 관련성 파악
- 산점도에 표시되는 각 점들은 자료의 관측값을 의미하고, 각 점의 위치는 관측값이 가지는 x축, y축 변수의 값으로 결정됨

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2

plt.scatter(x, y, s = area, c = colors, alpha = 0.5)
plt.show()
```

# 히스토그램

- 변수가 하나인 데이터의 빈도수를 막대 모양으로 나타낼 때 사용
- 통계분석에서 히스토그램은 가장 많이 사용되는 도구로, 데이터의 분석 및 분포를 파악하는 역할을 함
- x축에는 계급, y축에는 해당 도수 및 비율(개수)을 지정
    - 계급은 변수의 구간을 의미
    - 서로 겹치지 않아야 하고 계급들은 서로 붙어 있어야 함

# 파이차트

```python
ratio = [34, 32, 16, 18]
labels = ["Apple", "Banana", "Melon", "Grapes"]
explode = [0.05, 0.05, 0.05, 0.05]

plt.pie(ratio, labels = labels, autopct = "%.1f%%", startangle = 90,
        counterclock = False, explode = explode, shadow = True)
# %를 붙여서 나오게 할려면 %% 문자열 포매팅 형식을 사용
# startangle = 90 90도 회전 숫자를 입력하면 숫자만큼 각도 이동
# counterclock = False 시계방향으로 돈다 기본값은 반시계방향
# explode를 사용하면 차트가 떨어짐
# shadow = True 음영효과

plt.show()
```

# 서브플롯

- plt.subplot(nrow, ncol, pos) : 줄(행), 칸(열), 위치

```python
x = np.arange(1, 10)
y1 = x * 5
y2 = x * 1

plt.subplot(2, 1, 1) # 2행 1열로 나눈것 중에 첫번째 
plt.plot(x, y1)

plt.subplot(2, 1, 2) # 2행 1열로 나눈것 중에 두번째 
plt.plot(x, y2)

plt.show()
```

```python
plt.subplot(1, 2, 1) # 1행 2열로 나눈것 중에 첫번째 
plt.plot(x, y1)

plt.subplot(1, 2, 2) # 1행 2열로 나눈것 중에 두번째 
plt.plot(x, y2)

plt.show()
```

```python
y3 = x * 0.3
y4 = x * 0.2

plt.subplot(2, 2, 1) # 2행 2열로 나눈것 중에 첫번째 
plt.plot(x, y1)

plt.subplot(2, 2, 2) # 2행 2열로 나눈것 중에 두번째 
plt.plot(x, y2)

plt.subplot(2, 2, 3) # 2행 2열로 나눈것 중에 세번째 
plt.plot(x, y3)

plt.subplot(2, 2, 4) # 2행 2열로 나눈것 중에 네번째 
plt.plot(x, y4)

plt.show()
```

### 건강검진 데이터 남성과 여성의 허리둘레 히스토그램

```python
health.head()
```

```python
health["gender"] = health["gender"].map(lambda x: "Male" if x == 1 else "Female")
health["drinking"] = health["drinking"].map(lambda x: "Non-drinking" if x == 0 else "Drinking")
health["smoking"] = health["smoking"].map(lambda x: "Smoking" if x == 3 else "Non-smoking")
```

```python
health.head()
```

```python
male_data = health.loc[health["gender"] == "Male",
                      ["gender", "weight", "waist", "drinking", "smoking"]]

female_data = health.loc[health["gender"] == "Female",
                      ["gender", "weight", "waist", "drinking", "smoking"]]
```

```python
plt.figure(figsize = (10, 6))

sns.histplot(male_data["waist"], bins = 7, alpha = 0.5, label = "Male", kde = True)
sns.histplot(female_data["waist"], bins = 7, alpha = 0.5, label = "Male", kde = True,
            color = "r")

plt.xlabel("waist")
plt.ylabel("count")
plt.title("Waist Seaborn Histogram")
plt.legend()
plt.grid()
plt.show()
```

## 산점도 그래프 

### sns.regplot

```python
ax = plt.subplots()
ax = sns.regplot(x = "total_bill", y = "tip", data = tips, fit_reg = False)
# regplot 회귀선이 같이 그려짐
# fit_reg : 회귀선
# fit_reg = False를 하면 없어짐 기본값은 True
ax.set_title("Total Bill & Tip 산점도")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
plt.show()
```

## 산점도와 히스토그램 한 번에 그리기

### sns.jointplot

```python
joint = sns.jointplot(x = "total_bill", y = "tip", data =tips)
joint.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
joint.fig.suptitle("Joint Plot of Total Bill & Tip", fontsize = 10, y = 1.03)
plt.show()
```

- 산점도는 점이 겹칠 경우 구분하기 어려움 -> 육각 그래프
- 육각 그래프 : 2차원 표면에 육각형으로 데이터를 쌓아 표현, 특정 데이터의 개수가 많아지면 진한 색으로 표현

### kind = "hex"

- 육각 그래프로 변경

```python
joint = sns.jointplot(x = "total_bill", y = "tip", data =tips, kind = "hex")
joint.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
joint.fig.suptitle("Joint Plot of Total Bill & Tip", fontsize = 10, y = 1.03)
plt.show()
```

## 관계 그래프

### sns.pairplot

- 숫자로된 데이터를 가져와 그래프를 그려줌
- 컬럼수가 많으면 연산시간이 길어짐

```python
tips.head()
```

```python
fig = sns.pairplot(tips)
plt.show()
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot) # 산점도 그래프
pair_grid = pair_grid.map_diag(sns.histplot, kde = True) # 히스토그램
pair_grid = pair_grid.map_lower(sns.kdeplot) # 이차원 밀집도
plt.show()
```
