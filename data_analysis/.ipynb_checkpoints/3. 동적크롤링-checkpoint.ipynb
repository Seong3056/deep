## 원하는 태그를 tag_name으로 찾는 방법

```python
url = "https://pjt3591oo.github.io"

driver = webdriver.Chrome()

driver.get(url)

select_tag_p = driver.find_element(by = By.TAG_NAME, value = "p")
print(select_tag_p)
print(select_tag_p.tag_name)
print(select_tag_p.text)
```

```python
select_tag_p = driver.find_elements(by = By.TAG_NAME, value = "p")
print(select_tag_p)
```

## 원하는 태그를 class_name으로 찾는 방법

```python
url = "https://pjt3591oo.github.io"

driver = webdriver.Chrome()

driver.get(url)

select_class = driver.find_element(by = By.CLASS_NAME, value = "p")
print(select_class)
print(select_class.tag_name)
print(select_class.text)
```

```python
select_class = driver.find_elements(by = By.CLASS_NAME, value = "p")
print(select_class)
print(len(select_class))
```

## 원하는 태그를 css_selector로 찾는 방법

- bs4의 select() 와 동일

```python
url = "https://pjt3591oo.github.io"

driver = webdriver.Chrome()

driver.get(url)

select_css = driver.find_element(by = By.CSS_SELECTOR, value = "div.p")
print(select_css)
print(select_css.tag_name)
print(select_css.text)
```

```python
select_css = driver.find_elements(by = By.CSS_SELECTOR, value = "div.p")
print(select_css)
```

## 마우스 제어

```python
url = "https://pjt3591oo.github.io"

driver = webdriver.Chrome()

driver.get(url)

selected = driver.find_element(By.CSS_SELECTOR, "div.p > h3 > a")
print(selected.text)

selected.click()
```

```python
url = "https://pjt3591oo.github.io"

driver = webdriver.Chrome()

driver.get(url)

select_tags_a = driver.find_elements(By.CSS_SELECTOR, "a")

for i in select_tags_a:
    print(i.text,i.tag_name)
    i.click()
```

## selenium 활용

```python
url = "https://pjt3591oo.github.io/search"

driver = webdriver.Chrome()

driver.get(url)

search_box = driver.find_element(By.CSS_SELECTOR, "input#search-box")

search_box.send_keys("db")
search_box.send_keys(Keys.ENTER)

soup = BeautifulSoup(driver.page_source, "lxml")
items = soup.select("ul#search-results > li ")

for item in items:
    print(item)
```

```python
url = "https://pjt3591oo.github.io/search"

driver = webdriver.Chrome()

driver.get(url)

search_box = driver.find_element(By.CSS_SELECTOR, "input#search-box")

search_box.send_keys("db")
search_box.send_keys(Keys.ENTER)

soup = BeautifulSoup(driver.page_source, "lxml")
items = soup.select("ul#search-results > li ")

for item in items:
    title = item.select_one("h3").text
    description = item.select_one("p").text
    print(title)
    print(description)
```

#### 대기 방법 종류

- 1. time모듈 
    - 최후의 수단 밑에 두개 안 될 때 시도
- 2. 묵시적 대기
    - 페이지 정보를 다 받아 올 때 까지
- 3. 명시적 대기

# 예제 2. 네이버 지식백과

```python
url = "https://www.naver.com/"

driver = webdriver.Chrome()
driver.get(url) # 페이지 이동 

driver.implicitly_wait(3) # 묵시적 대기, 활성화를 최대 3초까지 기다린다

# 요소 찾기 - 검색창 찾고 키 전송
search = driver.find_element(By.CSS_SELECTOR, "input#query")
search.send_keys("고슴도치")
search.send_keys(Keys.ENTER) 

driver.implicitly_wait(3) # 페이지 이동 후 묵시적 대기 하는 것이 좋음

# 요소 찾기  - 지식백과에서 고슴도치 클릭
post = driver.find_element(By.CSS_SELECTOR, "a.area_text_title")
post.click()
```

# 예제3. 스타벅스

```python
url = "https://www.starbucks.co.kr/store/store_map.do?disp=locale"

driver = webdriver.Chrome()
driver.get(url) # 페이지 이동 

driver.implicitly_wait(3)

post = driver.find_elements(By.CSS_SELECTOR, " a.set_sido_cd_btn")
post[0].click()

time.sleep(3)

search = driver.find_elements(By.CSS_SELECTOR, "a.set_gugun_cd_btn")
search[13].click()

time.sleep(3)

soup = BeautifulSoup(driver.page_source, "lxml")
area = soup.select("li.quickResultLstCon > strong")
```

```python
[i.string for i in area]
```
